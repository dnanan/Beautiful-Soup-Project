```python
table = soup.find_all('table')[1]
```

```python
rows = table.find_all('tr')
```

```python
# This line of code is use to retrieve the link
base = "https://en.wikipedia.org/"
links = []
for row in rows[2:]:
  try:
    cell = row.find_all('td')[0]
    anchors = cell.find_all('a')[0]
    link = base + anchors['href']
    links.append(link)
    print(link)
  except:
    print("error")
```

```python
# This line is for retrieving the name
namedate = []
for row in rows[2:]:
  try:
    name  = row.find_all('td')[0].text.strip()
    date = row.find_all('th')[0].text.strip()
    namedate.append({
        "Name":name,
        "date":date
    })
    print(name)
  except:
    print("error")
```

```python
# This function retrieves the latitude
def GetLat(geo):
  latitude = re.split(" ",geo)[0]
  direction = latitude[-1]
  if direction == 'N':
    return float(latitude[:-2])
  else:
    return -float(latitude[:-2])
```

```python
# This function retrieves the longitude
def GetLon(geo):
  latitude = re.split(" ",geo)[1]
  direction = latitude[-1]
  if direction == 'E':
    return float(latitude[:-2])
  else:
    return -float(latitude[:-2])
```

```python
# This function is for retrieving the latitude and longitude together
def GetCoord(url):
  page = requests.get(url)
  soup = BeautifulSoup(page.content,'html.parser')
  geo = soup.find_all("span",class_="geo-dec")[0].text
  latitude = GetLat(geo)
  longitude = GetLon(geo)
  return(latitude,longitude)
```

```python
# This function retrieves the link
def GetName(url):
  name = re.split("/",url)[-1]
  name = re.sub("_"," ",name)
  return(name)
```

```python
def processList(urlList):
  Cities = []
  count = 1
  for url in urlList:
    try:
      coords = GetCoord(url)
      latitude = coords[0]
      longitude = coords[1]
      Cities.append({
          "latitude":latitude,
          "longitude":longitude
      })
      print(count,url,"GOOD")
    except:
      print(count,url,"BAD")
    count = count+1
  return(pd.DataFrame(Cities))
```
